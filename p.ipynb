```python
import pandas as pd

df = pd.read_excel("attendance.xlsx")
```

```python
df
```

```python
session_cols = [col for col in df.columns if "Session_" in col]

df["Attendance_Count"] = df[session_cols].sum(axis=1)

df["Certification_Status"] = df["Attendance_Count"].apply(
    lambda x: "Certified" if x >= 32 else "Not Certified"
)
```

```python
certified_students = df[df["Certification_Status"] == "Certified"]

print(certified_students["Name"])
```

```python
not_certified_students = df[df["Certification_Status"] == "Not Certified"]
```

```python
print(not_certified_students)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["Student_ID"], df["Attendance_Count"])

plt.axhline(y=32, linestyle="--")

plt.xlabel("Student ID")
plt.ylabel("Attendance Count")
plt.title("Attendance vs Certification Threshold")

plt.show()

```
